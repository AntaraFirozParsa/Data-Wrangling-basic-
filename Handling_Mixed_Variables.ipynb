```python
import pandas as pd
import numpy as np

data = {
    "Cabin": ["A/5 21171", "C85", "STON/O2. 3101282", "C123", None, None, "E46", None, None],
    "Ticket": ["5", "PC 17599", "6", "113803", "373450", "330877", "17463", "349909", "347742"],
    "number": ["5", "3", "6", "3", "A", "2", "2", "5", "1"],
    "Survived": [0, 1, 1, 1, 0, 0, 0, 0, 1]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Convert numeric part
df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce')

# Extract categorical part (like 'A')
df['number_categorical'] = df['number'].where(~df['number'].str.isdigit(), np.nan) #Is this value ONLY digits?
df.head(10)
```

```python
#Extract number (e.g., 85 from C85)
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') #\d → digit (0–9), + → one or more digits
df['cabin_num'] = pd.to_numeric(df['cabin_num'], errors='coerce')
#Extract category (e.g., C from C85)
df['cabin_cat'] = df['Cabin'].str.extract('([A-Za-z]+)') #[A-Za-z] → letters only
df.head(10)
```

```python
df['ticket_num'] = df['Ticket'].str.extract('(\d+)')
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce')

df['ticket_cat'] = df['Ticket'].str.extract('([A-Za-z/\.]+)')
```
