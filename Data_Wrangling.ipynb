# **Create the Messy Dataset**

```python
import pandas as pd
import numpy as np

data = {
    'Order ID':     [101, 102, 103, 104, 105, 103, 106, 107],
    'Customer Name':['Alice', 'bob', 'Charlie', 'DIANA', 'eve', 'Charlie', None, 'Grace'],
    'Product ':     ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Laptop', 'Keyboard', 'Mouse', 'Webcam'],
    ' Price(USD)':  ['$1200.00', '$25.50', '$75.00', '$300', '1100.00', '$75.00', '$25.50', None],
    'Order Date':   ['2024-01-15', '2024/01/18', '2024-02-15', '2024-01-20',
                     '2024-01-22', '2024-02-15', '2024-01-25', '2024-01-28'],
    'Quantity ':    [1, 3, 2, 1, 1, 2, 4, 1],
    'Category':     ['Electronics', 'electronics', 'Electronics', 'ELECTRONICS',
                     'Electronics', 'Electronics', 'electronics', 'Electronics'],
    'Status':       ['Delivered', 'Pending', 'Delivered', 'Shipped',
                     'Delivered', 'Delivered', 'pending', 'Shipped'],
}

df = pd.DataFrame(data)
print(df)
print("\nShape:", df.shape)
```

## **Audit the Mess**

```python
print("=== Data Types ===")
print(df.dtypes)

print("\n=== Missing Values ===")
print(df.isnull().sum())

print("\n=== Duplicates ===")
print("Duplicate rows:", df.duplicated().sum())
print("Duplicate Order IDs:", df.duplicated(subset=['Order ID']).sum())

print("\n=== Sample ===")
df.head(8)
```

## **Fix Column Names**

```python
# Strip whitespace and standardize all column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[^a-z0-9_]', '', regex=True)

print(df.columns.tolist())
# Expected: ['order_id', 'customer_name', 'product', 'priceusd', 'order_date', 'quantity', 'category', 'status']

# Rename the price column to something readable
df.rename(columns={'priceusd': 'price_usd'}, inplace=True)
df.head(8)
```

## **Remove Duplicates**

```python
print("Before:", df.shape)
df.drop_duplicates(subset=['order_id'], keep='first', inplace=True) #Keep the first occurrence of each duplicate
df.reset_index(drop=True, inplace=True)
print("After:", df.shape)
# Should go from 8 rows to 7 rows
```

## **Fix Data Types**

```python
# --- Price ---
df['price_usd'] = df['price_usd'].astype(str)   # make sure it's string
df['price_usd'] = df['price_usd'].str.replace('$', '')
df['price_usd'] = pd.to_numeric(df['price_usd'], errors='coerce') #pd.to_numeric() → converts text to numbers
                                                                  #errors='coerce' → if something cannot be converted, it becomes NaN
# --- Order Date ---
# Step 1: Convert everything to string and standardize separator
df['order_date'] = df['order_date'].astype(str).str.replace('/', '-')

# Step 2: Convert to datetime (handle mixed formats)
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', dayfirst=True)

 #Converts the column into datetime format --- filter dates, sort by time, extract year/month/day

# --- Quantity ---
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
df['quantity'] = df['quantity'].fillna(0) #Replaces missing values (NaN) with 0
df['quantity'] = df['quantity'].astype(int) #Converts values to integer

print(df.dtypes)
df.head(7)
```

## **Standardize Text Columns**

```python
# Title case for names and products
df['customer_name'] = df['customer_name'].str.strip().str.title()  #Converts text to Title Case (first letter capital)
df['product'] = df['product'].str.strip().str.title()

# Consistent casing for category and status
df['category'] = df['category'].str.strip().str.capitalize() #Only first letter uppercase, rest lowercase
df['status'] = df['status'].str.strip().str.capitalize()

print(df[['customer_name', 'product', 'category', 'status']])
```

### **Handle Missing Values**

```python
df.head(7)
# Check what's still missing
print(df.isnull().sum())

# Price: fill with median (robust to outliers)
df['price_usd'].fillna(df['price_usd'].median(), inplace=True)

# Customer name: fill with 'Unknown'
df['customer_name'].fillna('Unknown', inplace=True)

print("\nMissing after fix:", df.isnull().sum().sum()) #.sum().sum() → total missing values in whole dataset

df.head(7)
```

## **Feature Engineering**

```python
# Revenue column
df['revenue'] = df['price_usd'] * df['quantity']

# Extract date parts
df['order_month'] = df['order_date'].dt.month_name()
df['order_day'] = df['order_date'].dt.day_name()

# Order size category
df['order_size'] = pd.cut(df['quantity'], bins=[0, 1, 3, np.inf],
                           labels=['Small', 'Medium', 'Large'])

print(df[['order_id', 'product', 'revenue', 'order_month', 'order_size']])
```

## **Final Validation**

```python
print("=== Final Shape ===", df.shape)
print("\n=== Data Types ===\n", df.dtypes)
print("\n=== Missing Values ===\n", df.isnull().sum())
print("\n=== Clean Dataset ===")
df.head(10)
```
